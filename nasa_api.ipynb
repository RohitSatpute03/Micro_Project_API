```python
import requests

api_key = "DEMO_KEY"
api_url = "https://api.nasa.gov/mars-photos/api/v1/rovers/curiosity/photos"

params = {'sol': "1000", 'api_key': api_key}

response = requests.get(api_url, params=params)
print(response.status_code)

if response.status_code == 200:
    photos = response.json()
    print(photos)
else:
    print("Error:", response.json())
```

```python
data = response.json()

photos_data = data.get('photos')

for photo_data in photos_data:
    print(photo_data['camera']['full_name'])
```

```python
import requests
from skimage import io
import matplotlib.pyplot as plt

# Assume response contains the data from the previous API call
data = response.json()
photos_data = data.get('photos')

# Fetch and display the 5th image
for photo_data in photos_data[4:5]:
    img_url = photo_data['img_src']
    # Fetch image data with SSL verification disabled
    img_response = requests.get(img_url, verify=False)
    if img_response.status_code == 200:
        img = io.imread(img_response.content, plugin='imageio')
        io.imshow(img)
        plt.show()  # Display the image
        print(photo_data['camera']['full_name'])
    else:
        print("Failed to retrieve image")
```
